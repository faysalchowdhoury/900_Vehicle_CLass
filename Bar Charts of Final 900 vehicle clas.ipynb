```python
import pandas as pd

# Replace 'your_file.xlsx' with the actual file name
file_path = 'new_graph1.xlsx'

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(file_path)

# Melt the DataFrame to move rows to a new column
df_melted = df.melt(value_vars=df.columns[1:], var_name="Vehicle Class Number", value_name="Number of Vehicles")

# Print the first 5 rows of the melted DataFrame
print(df_melted.head())
```

```python
# Save the melted DataFrame to an Excel file
output_file_path = 'melted_data.xlsx'
df_melted.to_excel(output_file_path, index=False)

print("Melted DataFrame saved to Excel file:", output_file_path)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'melted_data.xlsx' with the actual file name
file_path = 'melted_data.xlsx'

# Read the Excel file into a Pandas DataFrame
df_melted = pd.read_excel(file_path)

# Create a bar plot using Seaborn
plt.figure(figsize=(10, 6))  # Adjust the figure size
sns.barplot(x="Vehicle Class Number", y="Number of Vehicles", data=df_melted)
plt.title("Number of Vehicles per Vehicle Class")
plt.xlabel("Vehicle Class Number")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45)  # Adjust rotation angle for x-axis labels
plt.tight_layout()  # Adjust layout for better visualization
plt.show()
```
